```python
if not cs:
  horse_features = pd.read_csv(os.path.join(local_paths.FEATURES_DIR, f'horse_features_{y}.csv'),index_col=0,  sep="\t")
  jockey_features = pd.read_csv(os.path.join(local_paths.FEATURES_DIR, f'jockey_features_{y}.csv'),index_col=0,  sep="\t", dtype={'jockey_id': str})
  results = pd.read_csv(os.path.join(local_paths.PREPROCESSED_DIR, f'results_{y}.csv')\
                        , index_col=0,  sep="\t", dtype={'jockey_id': str, 'trainer_id': str, 'owner_id': str})
  race_info = pd.read_csv(os.path.join(local_paths.PREPROCESSED_DIR, f'race_info_{y}.csv'), index_col=0,  sep="\t")

else:
  horse_features = pd.read_csv(os.path.join(local_paths.FEATURES_CS_DIR, f'horse_features_{y}.csv'),index_col=0,  sep="\t")
  jockey_features = pd.read_csv(os.path.join(local_paths.FEATURES_CS_DIR, f'jockey_features_{y}.csv'),index_col=0,  sep="\t", dtype={'jockey_id': str})
  results = pd.read_csv(os.path.join(local_paths.PREPROCESSED_CS_DIR, f'results_{y}.csv')\
                        , index_col=0,  sep="\t", dtype={'jockey_id': str, 'trainer_id': str, 'owner_id': str})
  race_info = pd.read_csv(os.path.join(local_paths.PREPROCESSED_CS_DIR, f'race_info_{y}.csv'), index_col=0,  sep="\t")
```

```python
race_info
```

```python
horse_features.columns
```

```python
results_add_info = results.merge(race_info, on='race_id', how='left')
```

```python
results_add_info
```

```python
results_add_horse_features = results_add_info.merge(
  horse_features,
  left_on=['horse_id', 'date'],
  right_on=['horse_id', 'reference_date'],
  how='left'
)
```

```python
results_add_horse_features['date'] = pd.to_datetime(results_add_horse_features['date'], errors='coerce')
results_add_horse_features['year'] = results_add_horse_features['date'].dt.year
```

```python
features = results_add_horse_features.merge(jockey_features, left_on=['jockey_id', 'year'], right_on=['jockey_id', 'reference_year'], how='left')
```

```python
# jockeyに関連するデータの欠損値を0で埋める
features.loc[:, features.columns.str.contains('jockey', case=False)] = features.loc[:, features.columns.str.contains('jockey', case=False)].fillna(0)
```

```python
features
```

```python
if not cs:
  features.to_csv(os.path.join(local_paths.FEATURES_DIR, f'features_{y}.csv'), sep="\t")
else:
  features.to_csv(os.path.join(local_paths.FEATURES_CS_DIR, f'features_{y}.csv'), sep="\t")
```
